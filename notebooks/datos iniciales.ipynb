```python
import geopandas as gpd

ruta = r"C:\Users\kenpa\OneDrive\Desktop\FORMACION\colombia-mercado-inmobiliario_v2\valor_ref_2020"

gdf = gpd.read_file(ruta)

print("=" * 50)
print("INFORMACIÓN DEL SHAPEFILE")
print("=" * 50)

# CRS
print("\nCRS:")
print(gdf.crs)

# EPSG
print("\nEPSG:")
print(gdf.crs.to_epsg())

# Campos
print("\nCAMPOS:")
for col in gdf.columns:
    print(f"- {col}")

# Tipos de datos
print("\nTIPOS DE DATOS:")
print(gdf.dtypes)

# Geometría
print("\nTIPOS DE GEOMETRÍA:")
print(gdf.geom_type.unique())

# Cantidad de registros
print("\nNÚMERO DE REGISTROS:")
print(len(gdf))

# Extensión espacial
print("\nBOUNDING BOX:")
print(gdf.total_bounds)

print("\nINFO GENERAL:")
print(gdf.info())
```

```python
import geopandas as gpd
import os

carpeta_base = r"C:\Users\kenpa\OneDrive\Desktop\FORMACION\colombia-mercado-inmobiliario_v2\data"

for subcarpeta in sorted(os.listdir(carpeta_base)):
    ruta_sub = os.path.join(carpeta_base, subcarpeta)
    if os.path.isdir(ruta_sub):
        for archivo in os.listdir(ruta_sub):
            if archivo.endswith(".shp"):
                gdf = gpd.read_file(os.path.join(ruta_sub, archivo))
                print(f"\n{'='*40}")
                print(f"Archivo: {subcarpeta}/{archivo}")
                print(f"Columnas: {list(gdf.columns)}")
                print(f"Registros: {len(gdf)}")
```
